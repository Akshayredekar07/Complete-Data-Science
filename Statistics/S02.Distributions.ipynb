### **Distributions**  
Distributions represent how data points are spread across a range of values. They provide insights into the frequency or probability of occurrences within a dataset.

#### 1. **Gaussian/Normal Distribution**  
- The most widely used distribution in statistics and data analysis.  
- Symmetrical, bell-shaped curve centered around the mean.  
- **Graph Example:** The bell curve showing mean height distribution in a population.  
- **Key Features:**  
  - Mean = Median = Mode.  
  - Follows the empirical rule: 68% of values within 1 SD, 95% within 2 SDs, 99.7% within 3 SDs.

#### 2. **Pre-computers**  
- Historical significance of Gaussian distribution, used even before the era of computational tools.  
- Example: Early statistical studies on population data by Carl Friedrich Gauss.  
- **Graph Example:** Heights of soldiers in the 19th century, represented as a bell curve.  


#### 3. **Probability Density Function (PDF)**  
- Mathematical representation of the probability of a value occurring in a continuous random variable.  
- Integral of the PDF over an interval gives the probability of the variable falling within that range.  
- **Graph Example:** A curve showing the likelihood of specific height ranges in a population.  


#### 4. **Heights of People**  
- A real-world example where data often forms a Gaussian distribution.  
- **Graph Example:** Height distribution of a population where most values cluster near the average, with fewer extreme tall or short individuals.




  The distribution of tuna fish lengths is modeled as $X \sim N(\mu = 75 \, \text{cm}, \sigma = 30 \, \text{cm})$. This means:  
  - Most tuna are around 75 cm in length.  
  - Lengths deviate by 30 cm on average.  
 
- **Question:**  
  What percentage of tuna are longer than 100 cm?  
 
- **Calculation:**  
  - Standardize $X = 100$:  
    $$
    Z = \frac{X - \mu}{\sigma} = \frac{100 - 75}{30} = 0.83
    $$  
  - Lookup $Z = 0.83$ in a standard normal distribution table:  
    $P(X < 100) = 0.7967$.  
  - Calculate the probability of lengths exceeding 100 cm:  
    $$
    P(X > 100) = 1 - 0.7967 = 0.2033 \, (20.33\%).
    $$  
 

 
 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Data for Gaussian Distribution
mu = 75  # Mean
sigma = 30  # Standard Deviation
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  # Range of values
y = norm.pdf(x, mu, sigma)  # Gaussian PDF

# Plotting the Gaussian Distribution for Tuna Lengths
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Gaussian Distribution', color='blue')
plt.axvline(100, color='red', linestyle='--', label='Length = 100cm (20.33% > 100cm)')
plt.axvline(mu, color='green', linestyle='--', label='Mean = 75cm')
plt.fill_between(x, y, where=(x > 100), color='red', alpha=0.3, label='Area > 100cm')
plt.title("Gaussian Distribution: Tuna Lengths", fontsize=16)
plt.xlabel("Length of Tuna (cm)", fontsize=14)
plt.ylabel("Probability Density", fontsize=14)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

#### 3. **Why Do Natural Phenomena Follow Gaussian Distribution?**  
 
1. **Genetic and Environmental Influences:**  
   - Genetic traits (like growth potential) and environmental factors (food, habitat) add up to produce a bell-shaped length distribution in tuna.  
 
2. **Central Limit Theorem (CLT):**  
   - When many small, independent random variables combine, their sum or average approaches a Gaussian distribution. For example, the combined effect of minor growth variations leads to this pattern.  
 
3. **Biological Symmetry:**  
   - Biological systems often grow symmetrically around an optimal value (mean), leading to a normal distribution.  
 
4. **Random Variability in Populations:**  
   - Individual differences in size or traits naturally create data clusters near the average, tapering off symmetrically.  


#### **1. Introduction to Poisson Distribution**
 
The Poisson distribution is a discrete probability distribution that describes the probability of a given number of events occurring within a fixed interval of time or space, under the following conditions:
- **Independence:** Events occur independently of each other.
- **Constant Mean Rate ():** The average rate of occurrence is constant over time or space.
- **Non-overlapping Events:** The probability of two or more events occurring in an infinitesimally small interval is negligible.
 
#### **2. Probability Mass Function (PMF)**
 
The probability of observing exactly $ k $ events is given by the Poisson PMF:
 
$$P(X = k) = \frac{e^{-\lambda} \cdot \lambda^k}{k!}$$
 
#### **Components of the PMF:**
- $ X $: Random variable representing the number of events.
- $ k $: Actual number of observed events ($ k = 0, 1, 2, \dots $).
- $ \lambda $: Average rate of events in the interval (mean).
- $ e $: Eulers number ($ e \approx 2.71828 $).
- $ k! $: Factorial of $ k $ ($ k! = k \cdot (k-1) \cdot (k-2) \cdot \dots \cdot 1 $).
 
#### **Example PMF Calculation:**
If a call center receives an average of 5 calls per minute ($ \lambda = 5 $), the probability of receiving exactly 3 calls in a minute is:
 
$$P(X = 3) = \frac{e^{-5} \cdot 5^3}{3!}$$
 
- $ e^{-5} \approx 0.0067 $
- $ 5^3 = 125 $
- $ 3! = 6 $
 
$$P(X = 3) = \frac{0.0067 \cdot 125}{6} \approx 0.1404$$
 
There is approximately a **14.04% chance** of receiving exactly 3 calls in a minute.
 
#### **3. Key Properties of the Poisson Distribution**
 
1. **Mean and Variance:**  
    Both the mean and variance of the Poisson distribution are equal to $ \lambda $:
    $$\text{Mean} = \lambda, \quad \text{Variance} = \lambda$$
 
2. **Skewness:**  
    - For small $ \lambda $, the distribution is right-skewed.
    - As $ \lambda $ increases, the distribution becomes symmetric, resembling a normal distribution.
 
3. **Additivity:**  
    If $ X_1 \sim \text{Poisson}(\lambda_1) $ and $ X_2 \sim \text{Poisson}(\lambda_2) $, then:
    $$X_1 + X_2 \sim \text{Poisson}(\lambda_1 + \lambda_2)$$
 
#### **4. Cumulative Distribution Function (CDF)**
 
The CDF of the Poisson distribution gives the probability of observing $ k $ or fewer events:
 
$$P(X \leq k) = \sum_{i=0}^k \frac{e^{-\lambda} \cdot \lambda^i}{i!}$$
 
#### **Example CDF Calculation:**
If $ \lambda = 5 $ and you want to calculate $ P(X \leq 3) $, sum the PMF for $ k = 0, 1, 2, 3 $:
 
$$P(X \leq 3) = P(X = 0) + P(X = 1) + P(X = 2) + P(X = 3)$$
 
- $ P(X = 0) = \frac{e^{-5} \cdot 5^0}{0!} = 0.0067 $
- $ P(X = 1) = \frac{e^{-5} \cdot 5^1}{1!} = 0.0337 $
- $ P(X = 2) = \frac{e^{-5} \cdot 5^2}{2!} = 0.0842 $
- $ P(X = 3) = 0.1404 $
 
$$P(X \leq 3) = 0.0067 + 0.0337 + 0.0842 + 0.1404 = 0.2650$$
 
Thus, there is a **26.5% chance** of observing 3 or fewer calls in a minute.
 
#### **5. Relationship to Other Distributions**
 
#### **a. Binomial Distribution:**
The Poisson distribution can be derived as a limiting case of the Binomial distribution when:
- The number of trials ($ n $) approaches infinity.
- The probability of success ($ p $) approaches 0.
- The product $ n \cdot p = \lambda $ remains constant.
 
Mathematically:
$$\text{Binomial}(n, p) \to \text{Poisson}(\lambda) \quad \text{as } n \to \infty, p \to 0$$
 
#### **b. Normal Distribution:**
For large $ \lambda $, the Poisson distribution approximates a normal distribution:
$$X \sim \text{Poisson}(\lambda) \approx \text{Normal}(\mu = \lambda, \sigma^2 = \lambda)$$
 
#### **6. Applications of Poisson Distribution**
 
1. **Queueing Theory:**  
    Modeling the arrival of customers at a service center, such as calls to a call center or customers in a queue.
 
2. **Reliability Engineering:**  
    Estimating the number of failures in a machine or system over a given period.
 
3. **Biology:**  
    Counting the number of mutations in a strand of DNA.
 
4. **Traffic Engineering:**  
    Modeling the number of cars passing through an intersection.
 
5. **Finance:**  
    Estimating the number of claims filed in insurance or defaults on loans.
 
#### **7. Visualization of Poisson Distribution**
 
Below is a summary of the behavior of the Poisson distribution for different values of $ \lambda $:
 
| $ \lambda $ | Description                              | Shape of Distribution           |
|---------------|------------------------------------------|----------------------------------|
| $ \lambda = 1 $ | Low event rate, highly right-skewed.   | Peak near 0, few larger values. |
| $ \lambda = 4 $ | Moderate event rate, less skewed.      | Peak near 4.                    |
| $ \lambda = 10 $ | High event rate, symmetric.           | Resembles normal distribution.  |
 
#### **Graphical Representation (Example):**
- **Small $ \lambda $:** Peaks at 0 and quickly declines.
- **Large $ \lambda $:** Becomes bell-shaped and symmetric.
 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the parameter 
lambda_val = 5

# Define range of k values
k_values = np.arange(0, 20)

# Calculate PMF
pmf_values = poisson.pmf(k_values, mu=lambda_val)

# Plot the distribution
plt.bar(k_values, pmf_values, color='blue', alpha=0.7)
plt.title('Poisson Distribution ( = 5)')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability')
plt.show()
```

### **1. Exponential Distribution PDF**:
The probability density function (PDF) of the Exponential distribution is given by:
$$
f(t) = \lambda e^{-\lambda t}, \quad t \geq 0
$$
Where:
- $t$: Time between events.
- $\lambda$: Event rate (the rate at which events occur).
 

For an Exponential distribution with $\lambda = 3$ (e.g., 3 arrivals per hour), the probability of waiting more than $t$ hours is:
$$
P(T > t) = e^{-\lambda t}
$$
For $t = 0.5$ hours, this becomes:
$$
P(T > 0.5) = e^{-3 \cdot 0.5} = e^{-1.5} \approx 0.2231
$$
This means there is approximately a 22.31% chance of waiting more than 30 minutes for the next customer in a queue.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, expon

# Parameters
lambda_rate = 4  # Event rate

# Poisson Distribution
x_poisson = np.arange(0, 15)  # Possible event counts
y_poisson = poisson.pmf(x_poisson, lambda_rate)  # PMF

# Exponential Distribution
x_exponential = np.linspace(0, 2, 500)  # Time intervals
y_exponential = expon.pdf(x_exponential, scale=1/lambda_rate)  # PDF

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Poisson Plot
axs[0].bar(x_poisson, y_poisson, color='skyblue', alpha=0.7, edgecolor='blue')
axs[0].set_title('Poisson Distribution ($\lambda=4$)')
axs[0].set_xlabel('Number of Events (k)')
axs[0].set_ylabel('Probability')
axs[0].grid(True, linestyle='--', alpha=0.5)

# Exponential Plot
axs[1].plot(x_exponential, y_exponential, color='red', label='Exponential PDF')
axs[1].fill_between(x_exponential, y_exponential, color='red', alpha=0.3)
axs[1].set_title('Exponential Distribution ($\lambda=4$)')
axs[1].set_xlabel('Time Between Events (t)')
axs[1].set_ylabel('Density')
axs[1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```


**Log-Normal Distribution Overview**

A random variable $X$ follows a **log-normal distribution** if its natural logarithm $Y = \log(X)$ follows a **normal distribution**. Mathematically:

- If $X \sim \text{Log-Normal}(\mu, \sigma^2)$, then:
  $$ Y = \log(X) \sim \mathcal{N}(\mu, \sigma^2) $$
  where $\mu$ is the mean and $\sigma^2$ is the variance of the underlying normal distribution of $Y$.

**Key Properties of Log-Normal Distribution**

- **Mean** of $X$:
  $$ \mathbb{E}[X] = e^{\mu + \frac{\sigma^2}{2}} $$
- **Variance** of $X$:
  $$ \text{Var}(X) = (e^{\sigma^2} - 1) \cdot e^{2\mu + \sigma^2} $$
- **Skewness**: Log-normal distributions are positively skewed, with a long right tail.

**Checking for Log-Normal Distribution**

1. **Log Transformation**: Transform the data $X$ by taking the natural logarithm:
   $$ Y = \log(X) $$
   If $X$ follows a log-normal distribution, $Y$ should follow a normal distribution.

2. **Visual Inspection**:
   - **Histogram**: Plot the histogram of $Y$. If it looks normal, the original data might follow a log-normal distribution.
   - **Q-Q Plot**: A Q-Q plot of $Y$ against a normal distribution helps check normality visually.

**Example: Zomato Delivery Times**

Assume you have Zomato delivery times:

$$ \text{Delivery times: } [12, 20, 35, 8, 45, 23, 50, 9, 60, 25] $$

1. **Log Transformation**:
   Apply the natural logarithm to the data:
   $$ Y = \log(\text{Delivery times}) = [2.48, 2.99, 3.56, 2.08, 3.81, 3.14, 3.91, 2.20, 4.09, 3.22] $$

2. **Visualize**:
   - **Histogram**: Plot the log-transformed values. Check if the distribution looks normal.
   - **Q-Q Plot**: Compare the transformed data to a normal distribution to confirm normality.



**Real-World Applications of Log-Normal Distribution**

- **Delivery Times**: Zomato delivery times, once log-transformed, can be modeled using a log-normal distribution.
- **Stock Prices**: Prices of stocks and financial returns often follow a log-normal distribution because their logarithms are normal.
- **Environmental Data**: Rainfall or wind speeds often exhibit log-normal behavior.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Zomato delivery times data
delivery_times = np.array([12, 20, 35, 8, 45, 23, 50, 9, 60, 25])

# Fit the log-normal distribution to the data
shape, loc, scale = lognorm.fit(delivery_times, floc=0)

# Generate a range of values for plotting
x = np.linspace(0, 70, 1000)

# Plot the PDF of the fitted log-normal distribution
plt.figure(figsize=(8, 6))
plt.plot(x, lognorm.pdf(x, shape, loc, scale), label='Fitted Log-Normal Distribution', color='blue')

# Plot histogram of the original data
plt.hist(delivery_times, bins=10, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Delivery Times Histogram')

# Add labels and title
plt.title('Log-Normal Distribution Fit to Zomato Delivery Times')
plt.xlabel('Delivery Time')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```


**In a factory that produces tyres, how many items need to be inspected before a defective tyre is detected? What distribution does this variable follow?**  

**Answer**:  
The number of inspections $X$ follows a **Geometric Distribution**:  
- $X \sim \text{Geometric}(P)$, where $P$ is the probability of a defective tyre.  
- Formula: $P(X = k) = (1 - P)^{k - 1} \cdot P$.


**If the number of weeks until the first network failure is modeled as $X$, what distribution does $X$ follow?**  

**Answer**:  
$X$ follows a **Geometric Distribution**:  
- $X \sim \text{Geometric}(P)$, where $P$ is the probability of failure in one week.  
- For continuous time, $X$ follows an **Exponential Distribution**:  
  $\Delta t \sim \text{Exponential}(\lambda)$, where $\lambda$ is the failure rate.


**What does the exponential distribution signify in relation to memorylessness, and how is it connected to the Poisson process?**  

**Answer**:  
The **Exponential Distribution** models the time between events in a Poisson process:  
1. **Memorylessness**:  
   $P(X > s + t \mid X > s) = P(X > t)$.  
2. **Poisson Relation**: Time intervals between events in a Poisson process follow $\text{Exponential}(\lambda)$, where $\lambda$ is the event rate.
