### Bayes' Theorem Explanation with Two Examples

Bayes' Theorem allows us to update the probability of an event (Event A) based on new evidence (Event B). This is useful when we have conditional probabilities, where the probability of an event depends on the occurrence of another event.



#### **Formula for Bayes' Theorem**:
$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$
Where:
- $ P(A|B) $: Posterior probability (probability of event A given B has occurred)
- $ P(B|A) $: Likelihood (probability of event B given A)
- $ P(A) $: Prior probability (initial probability of event A)
- $ P(B) $: Total probability of event B (overall probability of B)



### **Example 1: Medical Test (Disease Detection)**

Suppose you are trying to determine the probability that a patient has a disease (Event A) given that they tested positive (Event B).

- **Prior Probability**: $ P(A) = 0.01 $ (1% of people have the disease).
- **Likelihood**: $ P(B|A) = 0.9 $ (90% chance of a positive test if the patient has the disease).
- **False Positive Rate**: $ P(B|A') = 0.05 $ (5% chance of a positive test if the patient does not have the disease).

**Steps**:
1. **Calculate $ P(B) $** (the total probability of getting a positive test):
   $$
   P(A') = 1 - P(A) = 0.99 \quad \text{(Probability the patient doesn't have the disease)}
   $$
   Using the law of total probability:
   $$
   P(B) = P(B|A) \cdot P(A) + P(B|A') \cdot P(A') = (0.9 \cdot 0.01) + (0.05 \cdot 0.99) = 0.009 + 0.0495 = 0.0585
   $$
   
2. **Apply Bayes' Theorem** to find $ P(A|B) $ (the probability the patient has the disease given a positive test result):
   $$
   P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} = \frac{0.9 \cdot 0.01}{0.0585} \approx 0.154
   $$
   So, the probability that the patient has the disease given a positive test is approximately **15.4%**.



### **Example 2: Card Counting and Cheating Detection (Slot Machine)**

In this case, we want to determine the probability that a player is cheating (Event A) given that a slot machine shows "Yes" (Event B).

- **Prior Probability**: $ P(\text{Cheating}) = 0.05 $ (5% chance the player is cheating).
- **Likelihood**: $ P(\text{Machine = Yes | Cheating}) = 0.9 $ (90% chance the machine shows "Yes" if the player is cheating).
- **False Positive Rate**: $ P(\text{Machine = Yes | Not Cheating}) = 0.1 $ (10% chance the machine shows "Yes" if the player is not cheating).

#### **Bayesian Inference**

To calculate $ P(\text{Cheating | Machine = Yes}) $, we use Bayes' Theorem:

$$
P(\text{Cheating | Machine = Yes}) = \frac{P(\text{Machine = Yes | Cheating}) \cdot P(\text{Cheating})}{P(\text{Machine = Yes})}
$$

Where $ P(\text{Machine = Yes}) $ can be computed using the law of total probability:
$$
P(\text{Machine = Yes}) = P(\text{Machine = Yes | Cheating}) \cdot P(\text{Cheating}) + P(\text{Machine = Yes | Not Cheating}) \cdot P(\text{Not Cheating})
$$
Substitute the values:
$$
P(\text{Machine = Yes}) = (0.9 \cdot 0.05) + (0.1 \cdot 0.95) = 0.045 + 0.095 = 0.14
$$

Now apply Bayes' Theorem:
$$
P(\text{Cheating | Machine = Yes}) = \frac{0.9 \cdot 0.05}{0.14} \approx 0.3214
$$
So, the probability that the player is cheating given the machine shows "Yes" is approximately **32.14%**.



1. **Medical Test Example**:
   - We used Bayes' Theorem to determine the probability of a patient having a disease given a positive test result, which came out to **15.4%**.
   
2. **Cheating Detection in Slot Machines**:
   - We applied Bayes' Theorem to estimate the probability of cheating based on a machine showing "Yes," resulting in a **32.14%** chance of cheating.
