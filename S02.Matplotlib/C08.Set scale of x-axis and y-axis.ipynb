 set scale of x-axis and y-axis

```python
# Linear scaling 
import matplotlib.pyplot as plt 
import numpy as np 
a = np.arange(10000) 
b = np.arange(10000) 
plt.plot(a,b) 
plt.grid() 
plt.xscale('linear') 
plt.show()
```

```python
# Logarithmic scaling 
import matplotlib.pyplot as plt 
import numpy as np 
a = np.arange(10000) 
b = np.arange(10000) 
plt.plot(a,b) 
plt.grid() 
plt.xscale('log') 
plt.show() 
```

```python
# Logarithmic scaling on y-axis and Linear scaling on x-axis 
import matplotlib.pyplot as plt 
import numpy as np 
a = np.arange(10000) 
b = np.arange(10000) 
plt.plot(a,b) 
plt.grid() 
plt.xscale('linear') 
plt.yscale('log') 
plt.show() 
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
a = np.arange(10000) 
b = np.arange(10000) 
plt.plot(a,b) 
plt.grid() 
plt.xscale('log',base=2) #logarithmic scaling 
plt.yscale('log',base=9) #logarithmic scaling 
plt.show()
```
