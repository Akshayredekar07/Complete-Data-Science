```python
import pandas as pd

# Creating a sample DataFrame
df = pd.DataFrame({
    'Product': ['Laptop', 'Mobile', 'Tablet'],
    'Jan_Sales': [500, 700, 300],
    'Feb_Sales': [600, 750, 320],
    'Mar_Sales': [550, 720, 310]
})

print("Wide Format Data:")
print(df)
```

```python
df_melted = df.melt(id_vars=['Product'], var_name='Month', value_name='Sales')

print("\nLong Format Data:")
print(df_melted)
```

```python
df_melted['Month'] = df_melted['Month'].str.replace('_Sales', '')

print("\nCleaned Long Format Data:")
print(df_melted)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)
countries = [f"Country_{i}" for i in range(1, 51)]  
years = list(range(1900, 2021)) 

data = {
    'Country': np.repeat(countries, len(years)), 
}

for year in years:
    data[str(year)] = np.random.uniform(-2, 2, size=len(countries) * len(years))  

df = pd.DataFrame(data)
```

```python
print("Wide Format Climate Data:")
print(df.head())
```

```python
df_melted = df.melt(id_vars=['Country'], var_name='Year', value_name='Temperature_Anomaly')

# Convert Year column to integer
df_melted['Year'] = df_melted['Year'].astype(int)

print("\nLong Format Climate Data:")
print(df_melted.head())
```

```python
avg_temp_anomalies = df_melted.groupby('Year')['Temperature_Anomaly'].mean()

# Display first few years' average
print("\nAverage Global Temperature Anomalies:")
print(avg_temp_anomalies.head())
```

```python
hottest_year = avg_temp_anomalies.idxmax()
coldest_year = avg_temp_anomalies.idxmin()

print(f"\nHottest Year: {hottest_year} (Avg Anomaly: {avg_temp_anomalies.max():.2f}째C)")
print(f"Coldest Year: {coldest_year} (Avg Anomaly: {avg_temp_anomalies.min():.2f}째C)")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_temp_anomalies, marker='o', linewidth=2, color='red')
plt.axhline(0, linestyle='--', color='black', alpha=0.7)  # Baseline
plt.title("Global Average Temperature Anomalies (1900-2020)")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (째C)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_countries = df_melted[df_melted['Country'].isin(['Country_1', 'Country_2', 'Country_3'])]  # Select a few countries

sns.lineplot(data=top_countries, x='Year', y='Temperature_Anomaly', hue='Country', linewidth=2)
plt.title("Temperature Anomaly Trends for Selected Countries")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (째C)")
plt.legend(title="Country")
plt.grid(True)
plt.show()
```

```python
import pandas as pd

countries = np.array(["USA", "India", "Germany"])
years = np.arange(2000, 2005)  # 5 years

# Repeat each country for all years
repeated_countries = np.repeat(countries, len(years))

# Tile the years so they align correctly
repeated_years = np.tile(years, len(countries))

df = pd.DataFrame({"Country": repeated_countries, "Year": repeated_years})
print(df)
```
