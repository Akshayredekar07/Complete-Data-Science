```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 20, 1001)
fx = lambda x: x**2 -4*x+8
plt.figure(figsize=[5, 9],dpi=200)
plt.style.use('dark_background')

# Cambiar color y grosor de la gráfica de la función
plt.plot(x, fx(x), color='crimson', linewidth=4)

plt.grid('on')
ax = plt.gca()
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')

xk = 4
mu = 2 #0.2 y 5 iteraciones, 0.4 y 3 iteraciones, 2
yk = fx(xk)
plt.plot(xk, yk, 'ro')
plt.text(xk + 0.4, yk - 0.2, '$f(x_0)$', color='white')

for k in range(1, 3):
    x0 = xk
    y0 = fx(x0)
    xk = x0 - mu * (2 * x0 - 4)
    yk = fx(xk)
    print(f"{k}: xk = {xk}, yk ={yk}")
    # Cambiar color y grosor de la línea de actualización
    plt.plot([x0, xk], [y0, yk], 'ro--', color='gold', linewidth=2)
    plt.text(xk , yk - 0.3, '$f(x_' + str(k) + ')$', color='gold', va='center')

plt.grid()
plt.plot(2, 4, 'o', color='white', linewidth=4)
plt.savefig('imagen3.png')
plt.show()
```
