```python
df.drop(columns=['Remarks'], inplace=True)
```

```python
df.set_index('Sr No', inplace=True)
```

```python
df.rename(columns={
    'Date dd/mm/yyyy': 'date',
    'Startup Name': 'startup', 
    'Industry Vertical':'vertical',
    'SubVertical':'subvertical',
    'City  Location': 'city',
    'Investors Name': 'investor',
    'InvestmentnType': 'round',
    'Amount in USD': 'amount'},
    inplace=True
)
```

```python
df.columns
```

```python
df.head()
```

```python
df.info()
```

```python
df['amount'] = df['amount'].fillna(0)
```

```python
df['amount'] = df['amount'].str.replace(",", '')
df['amount'] = df['amount'].str.replace("undisclosed", '0')
df['amount'] = df['amount'].str.replace('unknown', '0')
df['amount'] = df['amount'].str.replace('Undisclosed', '0')
```

```python
df['amount'] = df['amount'].astype(str)  # Convert to string to handle NaN
df = df[df['amount'].str.isdigit().fillna(False)]  # Handle NaN values by replacing them with False
```

```python
df['amount'] = df['amount'].astype('float')
```

```python
df.info()
```

```python
df.head()
```

```python
def dollor_inr(dollor):
    inr = dollor*85.56
    return inr/10000000
```

```python
df['amount'] = df['amount'].apply(dollor_inr)
```

```python
df.head()
```

```python
df['date'] = df['date'].str.replace('05/072018', '05/07/2018')
```

```python
df['date'] = pd.to_datetime(df['date'],errors='coerce')
```

```python
df.head()
```

```python
df.info()
```

```python
df['date'].dt.month
```

```python
df =  df.dropna(subset=['date', 'startup', 'vertical', 'city', 'investor','round', 'amount'])
```

```python
df.info()
```

```python
df.to_csv('startup_cleanded.csv', index=False)
```
